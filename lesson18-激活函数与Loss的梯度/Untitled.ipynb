```python
import torch
import torch.nn.functional as F
```

```python
x = torch.randn(1,10)
```

```python
x
```

```python
w = torch.randn(10,1,requires_grad=True)
w
```

```python
o = torch.sigmoid(x@w)
o,o.shape
```

```python
loss = F.mse_loss(torch.ones(1,1),o)
```

```python
loss
```

```python
loss.backward()
```

```python
w.grad
```

验证公式是否与程序结果一致

```python
a = 0.1151
```

```python
2*(a-1)*a*(1-a)*1.2504
```

```python
w.grad[0]
```

可以看出公式与程序算出的结果是一致的
